###  <font face="courier"  ><center> <font color=#39ff14 > Web </font >scraping</center></font>

```python
import requests
import bs4
# importing requests and beautifulSoup
```

```python
raw = requests.get("https://www.ebay.com/sch/i.html?_from=R40&_trksid=p2380057.m570.l1313&_nkw=watches&_sacat=0")
raw.content
# getting all the data from the site using, but it isn't readable
```

```python
readable_soup = bs4.BeautifulSoup(raw.text,'lxml')
print(readable_soup.prettify())
# making the html data readable/understandable
```

```python
cases = readable_soup.find_all("li",class_="s-item s-item__pl-on-bottom")
print(len(cases))
```

```python
price = []
name = []
for i in cases:
    name_1 = i.find("h3",class_="s-item__title")
    name.append(name_1.string)
    price_1 = i.find("span",class_="s-item__price")
    price.append(price_1.string)
print(name)
print(price)
```

```python
import pandas as pd
dt=pd.DataFrame({'Name':name,'price':price})
print(dt)
```
